```python
from reportlab.pdfgen import canvas

def txt_to_pdf(txt_file, pdf_file):
    from reportlab.lib.pagesizes import A4
    from reportlab.pdfbase.pdfmetrics import stringWidth

    PAGE_WIDTH, PAGE_HEIGHT = A4
    LEFT_MARGIN = 40
    RIGHT_MARGIN = 40
    TOP_MARGIN = 40
    BOTTOM_MARGIN = 40
    FONT_NAME = "Helvetica"
    FONT_SIZE = 12
    LINE_SPACING = 15

    c = canvas.Canvas(pdf_file, pagesize=A4)
    c.setFont(FONT_NAME, FONT_SIZE)

    y = PAGE_HEIGHT - TOP_MARGIN
    max_width = PAGE_WIDTH - LEFT_MARGIN - RIGHT_MARGIN

    with open(txt_file, "r", encoding="utf-8") as f:
        for line in f:
            line = line.rstrip("\n")
            # Split line into chunks that fit the page width
            while line:
                # Find the max number of characters that fit
                for i in range(len(line), 0, -1):
                    chunk = line[:i]
                    width = stringWidth(chunk, FONT_NAME, FONT_SIZE)
                    if width <= max_width:
                        break
                c.drawString(LEFT_MARGIN, y, chunk)
                y -= LINE_SPACING
                if y < BOTTOM_MARGIN:
                    c.showPage()
                    c.setFont(FONT_NAME, FONT_SIZE)
                    y = PAGE_HEIGHT - TOP_MARGIN
                line = line[i:]
    c.save()

txt_to_pdf("..\kb\products\IQ SELF.txt", "..\kb\products\IQ SELF.pdf")
```
